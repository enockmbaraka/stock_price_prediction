```python
import yfinance as yf
import pandas as pd
```

```python


# Download historical stock data
tickers = ['GOOGL', 'META', 'AAPL', 'NVDA']
stock_data = yf.download(tickers, start="2021-10-30", end="2024-10-30", group_by="ticker")

# Check data for Google (GOOGL)
google_stock = stock_data['GOOGL']
meta_stock = stock_data['META']
aapl_stock = stock_data['AAPL']
nvda_stock = stock_data['NVDA']
print(google_stock.head())
```

```python
# Specify the file path where the CSV should be saved
file_path1 = 'google_stock_data.csv'
file_path2 = 'meta_stock_data.csv'
file_path3 = 'aapl_stock_data.csv'
file_path4 = 'nvda_stock_data.csv'
# Save the DataFrame to a CSV file
google_stock.to_csv(file_path1, index=True)
meta_stock.to_csv(file_path2, index=True)
aapl_stock.to_csv(file_path3, index=True)
nvda_stock.to_csv(file_path4, index=True)
```
